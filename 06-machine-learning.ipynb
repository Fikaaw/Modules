**Self-supervised Learning** 


“Sekarang kita sampai ke self-supervised learning. Kroma masih ingat kan supervised learning? Nah, jenis ini adalah supervised learning tanpa pelabelan oleh manusia atau tanpa keterlibatan manusia dalam tahapannya. Pada pembelajaran ini masih melibatkan label tetapi dihasilkan dari data input dan biasanya menggunakan algoritma heuristik,” paparku.

 

Aku memberitahu contoh dari self-supervised learning pada Kroma seperti autoencoder yang dilatih untuk menghasilkan output yang sama dengan inputnya. Contoh lain misalnya memprediksi frame selanjutnya pada sebuah video dengan diberikan frame sebelumnya, atau memprediksi kata selanjutnya pada sebuah teks dengan diberikan kata sebelumnya.

 

Aku juga mengingatkan Kroma bahwa sering terdapat kerancuan dalam mengategorikan self-supervised learning ke dalam supervised learning atau ke dalam unsupervised learning. Self-supervised learning dapat diinterpretasikan ulang sebagai supervised ataupun unsupervised learning, tergantung pada apakah pembuat memperhatikan mekanisme pembelajaran atau pada konteks penerapannya.

 

Di tengah penjelasanku, Kroma mendadak melempar sepotong pertanyaan.

 

“Waktu belajar sendiri pernah ketemu sama yang namanya semi-supervised learning­. Kalau jenis itu apakah sama dengan self-supervised ini?”

 

“Tidak, Kroma. Self-supervised learning berbeda dengan semi-supervised learning. Semi-supervised learning merupakan kombinasi dari supervised dan unsupervised learning, di mana terdapat sebagian kecil data berlabel dan sebagian besar data belum memiliki label,” jawabku.

 

**Teknik semi-supervised learning** 

secara khusus melibatkan langkah langkah berikut:

 

Melatih (training) model menggunakan data yang telah memiliki label (mirip seperti supervised learning) sampai model memberikan hasil yang baik.
Menggunakan model dengan melatih data yang belum memiliki label untuk memprediksi output.
Menghubungkan label dari data yang telah diberi label dengan label yang diprediksi tadi dan data input yang telah memiliki label dengan input data yang belum memiliki label.
Melatih model dengan cara yang sama sebagai satu kesatuan dari dataset yang telah diberi label.
 

Sembari Kroma menyimak catatanku, aku lanjut aku menjelaskan secara singkat perbedaan signifikan dari self-supervised learning dengan semi-supervised learning. Titik perbedaannya yaitu self-supervised learning tidak melibatkan data yang memiliki label, sedangkan semi-supervised learning masih melibatkan sejumlah kecil data yang telah memiliki label.

```python
# Mengimport library Pandas
import pandas as pd
pd.set_option("display.max_column", 10)

# Membaca dataset
dataset_credit_scoring = pd.read_excel('https://storage.googleapis.com/dqlab-dataset/credit_scoring_dqlab.xlsx')

# Membuat dataset
dataset = dataset_credit_scoring[['pendapatan_setahun_juta', 'kpr_aktif', 'durasi_pinjaman_bulan', 'jumlah_tanggungan', 'rata_rata_overdue', 'risk_rating']]

# Mengubah data kpr_aktif menjadi tipe integer: 'YA' = 1 dan 'TIDAK' = 0
dataset['kpr_aktif'] = dataset['kpr_aktif'].replace(['YA', 'TIDAK'], [1, 0])

# Mengubah data rata_rata_overdue menjadi numerik
mapping_dict = {
    "rata_rata_overdue": {
        "46 - 60 days": 60,
        "0 - 30 days": 30,
        "31 - 45 days": 45,
        "61 - 90 days": 90,
        "> 90 days": 91
    }
}
dataset = dataset.replace(mapping_dict)

# Menghapus kolom pendapatan_setahun_juta dan durasi_pinjaman_bulan
dataset = dataset.drop(['pendapatan_setahun_juta','durasi_pinjaman_bulan'], axis=1)

# Input variabel X dengan drop kolom risk_rating, karena kolom risk_rating digunakan sebagai lavel (y) (dependen)
X = dataset.drop('risk_rating',axis=1)

# Input variabel y dengan drop kolom label
y = dataset['risk_rating'].astype('category')

# Membagi data training dan data testing
from sklearn.model_selection import train_test_split

# Splitting data train 80% test 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Import library LogisticRegression
from sklearn.linear_model import LogisticRegression

# Membuat Objek model_lr
model_lr = LogisticRegression(solver='liblinear', random_state=0)

# Fitting model
model_lr.fit(X_train, y_train)

# Menjalankan data testing
y_predict = model_lr.predict(X_test)

# Menampilkan akurasi model
from sklearn.metrics import accuracy_score
print("Accuracy score: ", accuracy_score(y_test, y_predict))

import seaborn as sns
import matplotlib.pyplot as plt

# Melakukan evaluasi model dengan confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm.T, square=True, annot=True, cmap='Blues')
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.show()
```

```python
#Mengimport library Pandas
import pandas as pd
pd.set_option("display.max_column", 10)
#Membaca dataset
dataset_credit_scoring = pd.read_excel('https://storage.googleapis.com/dqlab-dataset/credit_scoring_dqlab.xlsx')
#Membuat dataset
dataset = dataset_credit_scoring[['pendapatan_setahun_juta', 'kpr_aktif', 'durasi_pinjaman_bulan', 'jumlah_tanggungan', 'rata_rata_overdue', 'risk_rating']]
#Mengubah data kpr_aktif menjadi tipe integer: 'YA' = 1 dan 'TIDAK' = 0
dataset['kpr_aktif'] = dataset['kpr_aktif'].replace(['YA', 'TIDAK'], [1, 0])
#Mengubah data rata_rata_overdue menjadi numerik
mapping_dict = {
	"rata_rata_overdue": {
		"46 - 60 days": 60,
		"0 - 30 days": 30,
		"31 - 45 days": 45,
		"61 - 90 days": 90,
		"> 90 days": 91
	}
}
dataset = dataset.replace(mapping_dict)
#Menghapus kolom pendapatan_setahun_juta dan durasi_pinjaman_bulan
dataset = dataset.drop(['pendapatan_setahun_juta','durasi_pinjaman_bulan'], axis=1)

#Input variabel X dengan drop kolom risk_rating, karena kolom risk_rating digunakan sebagai lavel (y) (dependen)
X = dataset.drop('risk_rating', axis=1)
#Input variabel y dengan drop kolom label
y = dataset['risk_rating'].astype('category')
#Membagi data training dan data testing
from sklearn.model_selection import train_test_split
#splitting data train 80% test 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Import library SVM
from sklearn import svm
#Membuat Objek model_svm
model_svm = svm.SVC(kernel='linear')
model_svm.fit(X_train, y_train)

#Menjalankan data testing
y_predict = model_svm.predict(X_test)

#Menampilkan akurasi model
from sklearn.metrics import accuracy_score
print("Accuracy score: ", accuracy_score(y_test, y_predict))

import seaborn as sns
import matplotlib.pyplot as plt

#Melakukan evaluasi model dengan confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm.T, square = True, annot = True, cmap = 'Blues')
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.show()
```
